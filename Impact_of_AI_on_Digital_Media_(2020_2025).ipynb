#**Exploratory Data Analysis (EDA)**

```python
#Univariate Analysis for the numerical columns
df.hist(figsize=(10,10))
plt.show()
```

The histograms display the distribution of key variables related to AI impact from 2020 to 2025. AI adoption rates and revenue increases are relatively high, with significant variability, while AI-generated content volumes show a more even distribution. Consumer trust and job loss due to AI also vary considerably, indicating diverse public perceptions and economic effects across regions or industries.

```python
#Bivariate Analysis for AI adoption vs. revenue increase
sns.scatterplot(x='AI Adoption Rate (%)', y='Revenue Increase Due to AI (%)', data=df)
plt.show()

```

The scatter plot shows a wide distribution of data points with no clear linear relationship between AI adoption rates and revenue increase. This suggests that other factors, aside from adoption rates, may be influencing the revenue outcomes in AI-driven industries.

```python
#Multivariate Analysis: Heatmaps for correlation among numerical variables
numerical_df = df.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numerical_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

The correlation heatmap reveals that the AI adoption rate has very weak correlations with most other variables, suggesting that adoption alone doesn't strongly affect factors like revenue or job loss. Notably, Job Loss Due to AI has a moderate positive correlation with Revenue Increase Due to AI (0.15). Human-AI Collaboration Rate and Consumer Trust in AI show very weak or negligible correlations with other variables, indicating a limited impact on the quantitative outcomes.

```python
#Categorical Analysis: Bar plots for distribution across industries and countries.

plt.figure(figsize=(12, 6))
sns.countplot(x='Industry', data=df, palette='Set2')
plt.title('Distribution of Industries')
plt.xticks(rotation=45)
plt.show()
```

The bar plot illustrates the distribution of industries in the dataset, with Media and Gaming standing out as the most represented sectors, followed by Legal and Automotive. The plot also shows more balanced representations across other industries such as Retail, Marketing, and Healthcare

##Data Interpretation & Analysis:
The analysis of the visualizations reveals several key trends. AI adoption rates vary significantly across industries, with media and gaming leading the way. Despite high adoption, consumer trust in AI remains low, suggesting concerns over AI's role in content creation.

Interestingly, a moderate correlation exists between job displacement and revenue increase, indicating that while AI adoption leads to job loss, it also boosts revenue, especially in sectors like automotive and media. Human-AI collaboration rates are high, but the impact on overall industry growth remains inconsistent. These insights underscore the complex relationship between AI integration and its outcomes.

#Recommendations
The recommendations:

**Foster Consumer Trust:** Industries with high AI adoption, such as media and gaming, should prioritize transparency and ethical AI practices to enhance consumer trust. Clear communication about AI's role in content creation can help mitigate skepticism.

**Balance Job Displacement with Retraining:** To address job losses due to AI, companies should invest in retraining programs to help employees transition to new roles that AI can't easily replace, such as AI maintenance or content moderation.

**Encourage Human-AI Collaboration:** Industries should focus on maximizing human-AI collaboration, leveraging AI tools to enhance productivity without fully replacing human creativity or decision-making.

**Adapt Regulations**Policymakers should craft regulations that strike a balance between fostering innovation and protecting workers and consumers. This includes guidelines for AI transparency, accountability, and data privacy.

**Monitor AI's Economic Impact**: Industries should track the economic impact of AI adoption to understand its long-term effects on revenue growth and market share, adjusting strategies accordingly.









**Conclusion**

AI is significantly transforming industries like media, gaming, and automotive, enhancing productivity and revenue while also presenting challenges like job displacement and consumer skepticism. The analysis reveals that while AI adoption is growing, especially in content-heavy sectors, trust remains a barrier to widespread acceptance. The correlation between AI adoption, job loss, and revenue growth suggests that businesses must strike a balance between technological advancement and workforce preservation. To ensure sustainable growth, industries must foster collaboration between humans and AI, invest in retraining programs, and advocate for transparent AI regulations that protect both workers and consumers.
